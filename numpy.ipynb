```python
import numpy as np

arr = np.array([1,3,5,])
print('numpy_array:', arr)
```

```python

arr2 = np.array([[1,3,5,],[4,6,8]])
print('numpy_array:', arr2)

arr2.shape
```

```python
#dtype

a = np.array([1,2,3])
print(a.dtype)
```

```python
#reshape


a = np.array([1,2,3, 4, 5, 6])
b = a.reshape(2,3)
print('reshaped-array', b)
```

```python
# astype

a = np.array([1,2,3])
b = a.astype(float)
print('reshaped-array', b)
```

```python
#np.zeros, np.ones

z = np.zeros((7, 4))
o = np.ones((4, 7))

print("zeros", z)
print("ones", o)
```

```python
#ndim


a = np.array([1,2,3])
print(a.ndim)
```

```python
#arrange


a = np.arange(1, 6)
print("arange", a)
```

```python
#sum(), mean(), max(), min()

a = np.array([1, 2, 3, 4, 5])

print('sum', a.sum())
print('mean', a.mean())
print('max', a.max())
print('min', a.min())
```

```python
#flatten
a = np.array([[1, 2 ,3], [1, 2 ,3]])

print('flatten', a.flatten())
```

```python
#t (transpose)
a = np.array([[1, 2],[3, 4]])

print('transpose', a.transpose())
```

```python
#dot
a = np.array([[1, 2],[3, 4]])
b = np.array([[1, 2],[3, 4]])

print('dot/multiply', a.dot(b))
```

```python
#random.rand()
a = np.random.rand(2, 3)
print('random_rand,', a)
```
